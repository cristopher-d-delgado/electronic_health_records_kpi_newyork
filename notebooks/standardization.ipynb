# Data Standardization for MySQL analysis

Purpose: Ensure the data types are in expected formats for MySQL. The data file is to large to open in Excel so Python will be used to format everything correctly.

```python
# import libraries 
import pandas as pd 
import numpy as np
```

```python
# Load data 
data = pd.read_csv('../data/Hospital_Inpatient_Discharges__SPARCS_De-Identified___2023.csv')
# Print data information 
print("Data Information:")
print(data.shape)
print(data.info())

# View values counts for each column
for column in data.columns:
    print(f"Value counts for column '{column}':")
    print(data[column].value_counts(dropna=False))
    print("\n")
```
